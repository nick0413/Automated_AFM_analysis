```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
import time

# Create data arrays
sin_array = np.sin(np.linspace(0, 4*np.pi, 100))
cos_array = np.cos(np.linspace(0, 2*np.pi, 100))

sin_x = np.linspace(0, 4*np.pi, 100)
cos_x = np.linspace(0, 2*np.pi, 100)

sin_df = pd.DataFrame({'x':sin_x, 'y': sin_array})
cos_df = pd.DataFrame({'x':cos_x, 'y': cos_array})

dfs = [sin_df, cos_df]

fig = go.Figure()

for df in dfs:
    fig.add_trace(go.Scatter(x=df['x'], y=df['y'], mode='lines', name='sin(x)'))

# Create a Plotly figure

# Add sin(x) trace

# Add cos(x) trace

# Customize layout
fig.update_layout(
    title='Sin and Cos',
    xaxis_title='x',
    yaxis_title='y'
)

# Show the figure
fig.show()
fig.data = []
```

```python
import plotly.graph_objs as go
import ipywidgets as widgets
from IPython.display import display
import numpy as np


# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

graphs = [y1, y2, y3]


# Create the Plotly FigureWidget
fig = go.FigureWidget()

# Add traces to the figure
trace1 = fig.add_scatter(x=x, y=y1, mode='lines', name='sin(x)')
trace2 = fig.add_scatter(x=x, y=y2, mode='lines', name='cos(x)')
trace3 = fig.add_scatter(x=x, y=y3, mode='lines', name='tan(x)')



# Widget to display selected graphs
output = widgets.Output()
display(output)

# List to store selected graphs
selected_graphs = []

# Function to handle clicks
def handle_click(trace, points, selector):
    with output:
        output.clear_output()  # Clear previous output
        if points.point_inds:  # Check if any point is clicked
            selected_curve_name = trace.name
            if selected_curve_name not in selected_graphs:
                selected_graphs.append(selected_curve_name)
            else:
                selected_graphs.remove(selected_curve_name)
    print('Graphs Selected: ', selected_graphs)
    return selected_graphs

# # Attach the click event handler to each trace using 'on_click'
# fig.data[0].on_click(handle_click)
# fig.data[1].on_click(handle_click)
# fig.data[2].on_click(handle_click)

# start = time.perf_counter()
submit_button = widgets.Button(description='Submit')

def on_submit(b):
    with output:
        print(f'Outlier Final Selection: {selected_graphs}')

# string = ''
for set in fig.data:
    set.on_click(handle_click)
    
submit_button.on_click(on_submit)
display(submit_button)

def user_input():
    user_confirm = input('Please input c to finish selecting graphs: ')
    return user_confirm

# Display the figure
display(fig)
```
