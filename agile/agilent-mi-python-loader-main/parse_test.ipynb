```python
import numpy as np
import matplotlib.pyplot as plt
from agilent_loader import load_mi
```

# IMAGE

# Load file

```python
mi = load_mi('paso6.mi')
```

# Access Data

```python
print("Number of buffers:", mi.buffer_count)
```

```python
print("Attributes in second buffer:", mi.buffers[1].properties)
print("Buffer unit of second buffer:", mi.buffers[1].bufferUnit)
```

```python
print("Attributes in file:", mi.properties)
```

```python
print("ServoIGain:", mi.servoIGain)
```

# Plotting example

```python
for buffer in mi.buffers:
    plt.imshow(buffer.data, origin='lower', extent=mi.extent)
    plt.title(buffer.bufferLabel + buffer.direction)
    plt.colorbar()
    plt.show()
```

# SPECTRO POINTS

```python
filename = "spectro_point.mi"
mi = load_mi(filename)
mi.bufferUnit
```

```python
buffer = mi.points[4].buffers[0]
measurement = buffer.approach

for curve in measurement.curves:
    z = measurement.z
    plt.plot(z, curve, 'r', alpha=0.5, lw=1)

measurement = buffer.retract

for curve in measurement.curves:
    z = measurement.z
    plt.plot(z, curve, 'k', alpha=0.5, lw=1)

plt.title(f"{buffer.bufferLabel} [{buffer.bufferUnit}]")
```
